```python
# get enron_df with only labeled emails
enron_df_labeled = enron_df[enron_df['label'].apply(lambda x: len(x) > 0 and x != [0, 0, 0])]

texts: pd.Series = enron_df_labeled['email_text']
labels: pd.DataFrame = enron_df_labeled['label']

# # concat texts and unlabeled_texts
# texts = pd.concat([texts, pd.Series(unlabeled_texts)], ignore_index=True)
# # labels should add to the end list [] for each text in unlabeled_texts
# labels= pd.concat([labels, pd.Series([[] for _ in range(len(unlabeled_texts))])], ignore_index=True)

```

```python
 
enron_df['split_labels'] = enron_df['violated_rules'].apply(lambda x: x.split(',') if x else [])
```

```python
print(len(enron_df[enron_df['split_labels'].apply(lambda x: '1.2' in x)]))
print(len(enron_df[enron_df['split_labels'].apply(lambda x: '1.3' in x)]))
print(len(enron_df[enron_df['split_labels'].apply(lambda x: '1.1' in x)]))


# get rows where split labels is empty and violated rules is not empty
enron_df[(enron_df['split_labels'].apply(lambda x: len(x) == 0)) & (enron_df['violated_rules'].apply(lambda x: len(x) > 0))].head(3)
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import MultiLabelBinarizer


# mlb = MultiLabelBinarizer()
# labels_encoded = mlb.fit_transform(labels)
mlb = MultiLabelBinarizer(classes=['legal', 'financial_data', 'business_or_financial'])
labels_encoded = mlb.fit_transform(labels)  # Ensure 'labels' contains examples from all three categories

# Convert texts to TF-IDF features
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels_encoded, test_size=0.2, random_state=42, stratify=labels_encoded)
```

```python
print(texts)
```

```python

from sklearn.ensemble import RandomForestClassifier
# Train a Random Forest model within a OneVsRestClassifier
model = OneVsRestClassifier(RandomForestClassifier(n_estimators=100))
# model = OneVsRestClassifier(RandomForestClassifier(n_estimators=100))
model.fit(X_train, y_train)
```

```python

# Predict and evaluate the model
predictions = model.predict(X_test)
print(f'prediction 1: {predictions[0]} with label {y_test[0]}')

print(y_test.shape)
```

```python

# Specify labels to ensure all classes are included in the report
print(classification_report(y_test, predictions, target_names=mlb.classes_, labels=np.arange(mlb.classes_.shape[0]), digits=5))
```

```python
# test on all labeled data
X_all = vectorizer.transform(texts)
y_all = labels_encoded
predictions_all = model.predict(X_all)
print(classification_report(y_all, predictions_all, target_names=mlb.classes_, labels=np.arange(mlb.classes_.shape[0]), digits=5))
```
